```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 0.2)

for i in range(1,6):
    plt.plot(x,x**i)

plt.show()
```

```python
x = np.linspace(-4*np.pi, 4*np.pi, 300)

sx = 4*np.sin(x)
cx = np.cos(x*2)

plt.plot(x, cx, label = "cos")
plt.plot(x, sx, label = "sin")
plt.legend(loc="upper left")
plt.show()
```

```python
r =10
c = (-2, 2)

ang = np.arange(0, 360, 1)
x = r*np.cos(np.deg2rad(ang))-c[0]
y = r*np.sin(np.deg2rad(ang))+c[1]

plt.figure(figsize=(5,5))
plt.plot(x,y,'o')
plt.show()
```

```python
image = np.zeros((100, 100))

for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        image[y, x] = np.sin(x)*np.cos(y)

        
plt.imshow(image, cmap ="RdPu")
plt.colorbar()
plt.show()
```

```python
image = np.zeros((100, 100))#       25
image[::2, ::2] =1
image[1::2, 1::2] =1

plt.imshow(image)
plt.colorbar()
plt.show()
```

```python
image = np.ones((600, 600))

x = np.arange(600).reshape(600, 1)
y = np.arange(600).reshape(1, 600)

mask = ((y - 300) ** 2 + ( x - 300) ** 2) < 100 ** 2

image[mask] = 0
plt.imshow(image)
plt.colorbar()          

plt.show()
```
